```python
import os
import sys

# Imposta la directory delle applicazioni OTB
otb_application_path = "/content/OTB/OTB_install/lib/otb/python"
os.environ["OTB_APPLICATION_PATH"] = otb_application_path

# Aggiungi il percorso per i moduli Python di OTB
otb_python_path = "/content/OTB/OTB_install/lib/otb/python"
sys.path.append(otb_python_path)

# Importa il modulo otbApplication dopo aver impostato il percorso
otb_prefix_path = "/content/OTB/OTB_install"
os.environ["CMAKE_PREFIX_PATH"] = otb_prefix_path
os.environ["OTB_APPLICATION_PATH"] = os.path.join(otb_prefix_path, "lib", "otb", "python")
os.environ["PATH"] = os.path.join(otb_prefix_path, "bin") + os.pathsep + os.environ["PATH"]
sys.path.insert(0, os.path.join(otb_prefix_path, "lib", "python"))
os.environ["LC_NUMERIC"] = "C"
os.environ["GDAL_DATA"] = os.path.join(otb_prefix_path, "share", "gdal")
os.environ["PROJ_LIB"] = os.path.join(otb_prefix_path, "share", "proj")
os.environ["GDAL_DRIVER_PATH"] = "disable"
os.environ["OTB_MAX_RAM_HINT"] = "1000"
```
